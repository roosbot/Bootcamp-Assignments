```python
# Library import
import pandas as pd
import numpy as np

from pandas.util import hash_pandas_object
```

```python
# Read customer data
cdata = pd.read_csv("./data/customers.csv")

# Read order data
odata = pd.read_csv("./data/orders.csv")

# Read product data
pdata = pd.read_csv("./data/products.csv")
```

```python
# Hash order data
odata["Email ID"] = hash_pandas_object(odata["Email (Billing)"])
odata["Order ID"] = hash_pandas_object(odata["Order Number"])

# Hash customer data
cdata["Email ID"] = hash_pandas_object(cdata["Email (Billing)"])

# Hash product data
pdata["Order ID"] = hash_pandas_object(pdata["Order Number"])
```

```python
# Drop columns
odata = odata.drop("Email (Billing)",axis=1)
odata = odata.drop("Order Number",axis=1)

cdata = cdata.drop("Email (Billing)",axis=1)

pdata = pdata.drop("Order Number",axis=1)
```

```python
# Export data to csv

odata.to_csv("./data/orders_hashed.csv", index=False)
cdata.to_csv("./data/customers_hashed.csv", index=False)
pdata.to_csv("./data/products_hashed.csv", index=False)
```
