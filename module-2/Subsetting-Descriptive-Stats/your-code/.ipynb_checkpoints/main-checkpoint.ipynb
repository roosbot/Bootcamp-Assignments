####  What is the average purchase price?

```python
# your code here

# Do you mean item price or average order value?

## average item price
overview = orders.describe().transpose()
avg_item_price = overview["mean"][8]
print("The average item price is", int(avg_item_price))

## create customer dataframe
customers = orders.groupby(["CustomerID","Country"],as_index=False).agg({"amount_spent":"sum","Quantity":"sum","InvoiceNo":"count"})
customers.columns=["CustomerID","Country","Ordervalue","Quantity","Orders"]
customers["Avg_ordervalue"] = customers["Ordervalue"]/customers["Orders"]

## average order value
overview2 = customers.describe().transpose()
avg_ordervalue = overview2["mean"][4]
print("The average order value is", int(avg_ordervalue))
```

#### What are the highest and lowest purchase prices? 

```python
## highest and lowest item price
min_purchase_price = overview["min"][8]
max_purchase_price = overview["max"][8]
print("The highest item price is", int(max_purchase_price))
print("The lowest item price is", int(min_purchase_price))

## highest and lowest order value
min_ordervalue = overview2["min"][1]
max_ordervalue = overview2["max"][1]
print("The highest order value is", int(max_ordervalue))
print("The lowest order value is", int(min_ordervalue))
```

#### Select all the customers who have bought more than 50 items.
**Hint**: Remember that you are not asked to find orders with more than 50 items but customers who bought more than 50 items. A customer with two orders of 30 items each should appear in the selection.

```python
# your code here
# select = customers_spain[(customers_spain["Orders"] > 50)]
# select

customers_group = orders.groupby("CustomerID").agg("sum")
customers_group

big_customers = customers_group[customers_group["Quantity"] > 50].reset_index()
big_customers[["CustomerID", "Quantity"]].head()
```
