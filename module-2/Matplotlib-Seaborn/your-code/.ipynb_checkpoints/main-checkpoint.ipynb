#### The data we will use in this challenge is:

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x, y) and (x, z).
There are 2 ways of doing this. Do in both ways.

**Hint**: Check out the nrows, ncols and index arguments of subplots. 

Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# Method 1 - (x,y)
plt.style.use('seaborn-darkgrid')

plt.plot(x, x, label = "x")
plt.plot(x, y, label = "x*2")

plt.xlabel("x label")
plt.xlabel("y label")

plt.title("x,y plot")
plt.legend()
plt.show()
```

```python
# Method 1 - (x,z)
plt.style.use('seaborn-darkgrid')

plt.plot(x, x, label = "x")
plt.plot(x, z, label = "x**2")

plt.xlabel("x label")
plt.xlabel("z label")

plt.title("x,z plot")
plt.legend()
plt.show()
```

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
# Method 2
plt.style.use('seaborn-darkgrid')

fig, [ax1,ax2] = plt.subplots(nrows=1,ncols=2,constrained_layout=True)
ax1.plot(x,y)
ax1.set_title("x,y plot")
ax1.set_xlabel("x label")
ax1.set_ylabel("y label")

ax2.plot(x,z)
ax2.set_title("x,z plot")
ax2.set_xlabel("x label")
ax2.set_ylabel("z label")

plt.show()
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
plt.style.use('seaborn-darkgrid')

fig, [ax1,ax2] = plt.subplots(nrows=1,ncols=2,constrained_layout=True, figsize=(8, 4))
ax1.plot(x,y)
ax1.set_title("x,y plot",fontname="Times New Roman", fontsize=18)
ax1.set_xlabel("x label",fontname="Times New Roman", fontsize=18)
ax1.set_ylabel("y label",fontname="Times New Roman", fontsize=18)

ax2.plot(x,z)
ax2.set_title("x,z plot",fontname="Arial", fontsize=18)
ax2.set_xlabel("x label",fontname="Arial", fontsize=18)
ax2.set_ylabel("z label",fontname="Arial", fontsize=18)
plt.show()


# Fontnames
## Arial
## Comic Sans MS
## sans-serif
## Resource: https://matplotlib.org/3.1.1/tutorials/text/text_props.html

# Default font for all graphs:
## The default sans-serif font is COMIC SANS
### matplotlib.rcParams['font.sans-serif'] = "Comic Sans MS"
## ALWAYS use sans-serif fonts
# matplotlib.rcParams['font.family'] = "sans-serif"
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
y1 = x**2
y2 = np.exp(x)

plt.style.use('seaborn-darkgrid')

fig, [ax1,ax2] = plt.subplots(nrows=1,ncols=2,constrained_layout=True, figsize=(8, 4))
ax1.plot(x,y1)
ax1.set_title("y = x2")
ax1.set_xlabel("x label")
ax1.set_ylabel("y label")
ax1.set_xscale('linear')
ax1.legend("y1")

ax2.plot(x,y2)
ax2.set_title("y = ex")
ax2.set_xlabel("x label")
ax2.set_ylabel("y label")
ax2.set_yscale('log')
ax2.legend("y2")

plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
plt.style.use('seaborn-darkgrid')

fitbit_stats = fitbit.groupby(["Months_encoded"])[["Steps"]].agg("median").reset_index()

x = fitbit_stats["Months_encoded"]
y = fitbit_stats["Steps"]

plt.scatter(x, y, label = "x")

plt.xlabel("Months")
plt.ylabel("Steps")

plt.title("Median steps per month")
plt.legend()
plt.show()

plt.savefig("Median steps per month.png")
```

```python
# MEDIAN STEPS BY WORK_OR_WEEKEND
fitbit_stats = fitbit.groupby(["Work_or_Weekend"])[["Steps"]].agg("median").reset_index()

x = fitbit_stats["Work_or_Weekend"]
y = fitbit_stats["Steps"]

plt.bar(x, y)
plt.xticks(x,("Work","Weekend"))
# plt.xlabel("Work or Weekend")
# plt.ylabel("Steps")

plt.title("Steps per Work or Weekend")
plt.show()

plt.savefig("Steps per Work or Weekend.png")
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
x = "Steps"
y = ["Minutes Lightly Active", "Minutes Very Active", "Minutes Sedentary"]

for s in range(3):
    sns_plot = sns.lmplot(x="Steps",y=y[s],data=fitbit,fit_reg=False)
    sns_plot.savefig(y[s] + " vs Steps.png")
```
