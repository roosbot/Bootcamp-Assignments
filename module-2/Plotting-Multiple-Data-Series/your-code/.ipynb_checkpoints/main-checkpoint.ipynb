# Plotting Multiple Data Series

Complete the following set of exercises to solidify your knowledge of plotting multiple data series with pandas, matplotlib, and seaborn. Part of the challenge that comes with plotting multiple data series is transforming the data into the form needed to visualize it like you want. For some of the exercises in this lab, you will need to transform the data into the form most appropriate for generating the visualization and then create the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import datetime

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('../data/liquor_store_sales.csv')
data.head()
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
selection = data[["ItemType","RetailSales","RetailTransfers","WarehouseSales"]].groupby(["ItemType"]).agg("sum")
selection.plot.bar()

plt.title("Total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type")
plt.xlabel("ItemType")
plt.ylabel("Sales")
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
# First calculate total sales
data["TotalSales"] = data["RetailSales"]+data["WarehouseSales"]
data = data.sort_values(by="TotalSales",ascending=False)
data.head()
```

```python
# Find list of top 10 suppliers
top_suppliers = data[["Supplier","ItemType","TotalSales"]].groupby(["Supplier"]).agg("sum").reset_index()
top_suppliers = top_suppliers.sort_values(by="TotalSales",ascending=False)
top_suppliers = top_suppliers[:10]
# top_suppliers = list(top_suppliers["Supplier"])
top_suppliers
```

```python
# Select data top 10 suppliers
selection = data[data["Supplier"].isin(top_suppliers["Supplier"])]
selection
```

```python
# Create new dataframe with total sales per supplier and item type
data_grouped = selection[["Supplier","ItemType","TotalSales"]].groupby(["Supplier","ItemType"]).agg("sum").reset_index()
data_grouped

```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
data.sort_values(by=["Year","Month"], inplace=True)
data_mean = data[["RetailSales","RetailTransfers","WarehouseSales","Year","Month"]].groupby(["Year","Month"]).agg("mean")

data_mean.plot()

plt.xlabel("Month")
plt.ylabel("Average Sales")

```

## 4. Plot the same information as above but as a bar chart.

```python
data_mean.plot.bar()
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
# Create date column
data["Date"] = pd.to_datetime(data[["Year","Month"]].assign(DAY=1))
data["Date"] = data["Date"].dt.strftime("%Y-%m")

data.head()
```

```python
# Crate pivot table
data_pivot = data.pivot_table(index=["Date"], columns=["ItemType"], values=["RetailSales"]
                              ,aggfunc={"RetailSales":"sum"} )
data_pivot
```

```python
# Create multiple line plot
data_pivot.plot(kind="line", figsize=(17, 10), rot=0)                                       
plt.title("Retail Sales summed by Item Type over time")
plt.xlabel("Date")
plt.ylabel("Sales")
```

## 6. Plot the same information as above but as a bar chart.

```python
data_pivot.plot.bar(width=1)
```
