# StackAPI

#### Import the necessary libraries here:

```python
from stackapi import StackAPI
SITE = StackAPI('stackoverflow')

# I used https://www.epochconverter.com/ to convert timestamp
# Handy documentation: https://stackapi.readthedocs.io/en/latest/user/advanced.html#advanced
```

#### Question 1: Find the questions and answers of last month.

```python
# Questions of last month (April)
questions = SITE.fetch('questions',fromdate=1585699200, todate=1588204800)
questions
```

```python
# Answers of last month (April)

answers = SITE.fetch('questions',fromdate=1585699200, todate=1588204800)
answers
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here
SITE = StackAPI('stackoverflow')
answers = SITE.fetch('answers',fromdate=1589932800,todate=1590019199, tagged='python', sort='votes')
answers
```

#### Question 3: Find the answers with id 6784 and 6473.

```python
# your code here
SITE = StackAPI('stackoverflow')
answers = SITE.fetch('answers',ids=[6784, 6473])
answers
```
